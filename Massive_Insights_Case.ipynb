```python
# categorize the food_items
order_items["Dinner"] = np.where((order_items["food_item"] == "Braised Beef Noodles") | (order_items["food_item"] == "Gam Ja Tang"),
                                1,0)

order_items["Snacks"] = np.where((order_items["food_item"] == "Pork Belly Bao") | (order_items["food_item"] == "Cha Siu Bao"),
                                1,0)

order_items["Beverages"] = np.where((order_items["food_item"] == "Herbal Grass Jelly Drink") | (order_items["food_item"] == "Passionfruit Green Tea") | (order_items["food_item"] == "Red Bean & Taro Slush"),
                                1,0)

order_items["Dessert"] = np.where((order_items["food_item"] == "Matcha Cheese Cake") | (order_items["food_item"] == "Black Sugar Bubble Tea"),
                                1,0)
```

```python
# condtion create 

conditions = [
    order_items["Dinner"] == 1,
    order_items["Snacks"] == 1,
    order_items["Beverages"] == 1,
    order_items["Dessert"] == 1
    ]

# Set the new column values in order of the conditions they should
#	correspond to:
choices = ['Dinner', 'Snacks', 'Beverages',"Dessert"]

# Add new column based on conditions and choices:
order_items['Food Category'] = np.select(conditions, choices)
```

```python
conn = sqlite3.connect('test_database')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS orders (order_id,ordered_at,delivered_at,delivered_to,order_channel)')
c.execute('CREATE TABLE IF NOT EXISTS order_items (item_id,order_id,food_item,amount_paid)')
conn.commit()
```

```python
order.to_sql('orders', conn, if_exists='replace', index = False)
order_items.to_sql('order_items', conn, if_exists='replace', index = False)
    
    
c.execute('''  
SELECT food_item, 
COUNT(DISTINCT order_items.order_id) /
COUNT(DISTINCT orders.delivered_to) as ReorderRate
FROM order_items
JOIN orders on
orders.order_id = order_items.order_id
GROUP BY 1
ORDER BY 2 desc
          ''')

for row in c.fetchall():
    print (row)
```
